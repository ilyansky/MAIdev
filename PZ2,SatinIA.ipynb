```python
# Метод половинного деления
# x ^ 3 - x, x принадлежит [0, 1]
def findRatio(fAlpha, fBeta, firstIteration, previousInterval):
    print(f'f_alpha = {fAlpha}, f_beta = {fBeta}')
    if not firstIteration:
        if previousInterval != 0.0:
            num = abs((fAlpha - fBeta) / previousInterval)
            print(f'Настоящий интервал / предыдущий интервал = {num}')    
    print()

a = 0
b = 1
fAlpha = 0
fBeta = 0
firstIteration = True
previousInterval = 0

eps = 0.05
sigma = 0.02

while (b - a) >= eps:
    previousInterval = fBeta - fAlpha
    
    alpha = ((a + b) / 2) - sigma
    fAlpha = alpha ** 3 - alpha
    beta = ((a + b) / 2) + sigma
    fBeta = beta ** 3 - beta
    if fAlpha < fBeta:
        b = beta
    else:
        a = alpha

    findRatio(fAlpha, fBeta, firstIteration, previousInterval)
    firstIteration = False
    
x = (a + b) / 2
f = x ** 3 - x
print(x, f)
```

```python
# Метод золотого сечения
# x ^ 3 - x, x принадлежит [0, 1]
a = 0
b = 1
fAlpha = 0
fBeta = 0
firstIteration = True
previousInterval = 0

eps = 0.1
lam = 0.61803

alpha = a + (1 - lam) * (b - a)
beta = a + lam * (b - a)

while (b - a) >= eps:
    previousInterval = fBeta - fAlpha
    
    fAlpha = alpha ** 3 - alpha
    fBeta = beta ** 3 - beta
    if fAlpha <= fBeta:
        b = beta
        beta = alpha
        alpha = a + (1 - lam) * (b - a)
    else:
        alpha = beta
        beta = a + lam * (b - a)
    findRatio(fAlpha, fBeta, firstIteration, previousInterval)
    firstIteration = False
x = (a + b) / 2
f = x ** 3 - x
print(x, f)
```

```python
# Метод Фибоначчи
# x ^ 3 - x, x принадлежит [0, 1]
a = 0
b = 1
fAlpha = 0
fBeta = 0
firstIteration = True
previousInterval = 0

eps = 0.1
sigma = 0.04

# (b - a) / eps = 10 => 0: 1, 1: 1, 2: 2, 3: 3, 4: 5, 5: 8, 6: 13 => 6 итераций
fib = [1, 1, 2, 3, 5, 8, 13]
k = 1
n = 6

alpha = a + (fib[n-2] / fib[n]) * (b - a)
beta = a + (fib[n-1] / fib[n]) * (b - a)

while k != n - 2:
    previousIteration = fBeta - fAlpha
    fAlpha = alpha ** 3 - alpha
    fBeta = beta ** 3 - beta
    findRatio(fAlpha, fBeta, firstIteration, previousIteration)
    firstIteration = False
    
    if fAlpha < fBeta:
        b = beta
        beta = alpha
        alpha = a + (fib[n - k - 2] / fib[n - k]) * (b - a) 
    else:
        a = alpha
        alpha = beta
        beta = a + (fib[n - k - 1] / fib[n - k + 1]) * (b - a) 
    k += 1


fAlpha = alpha ** 3 - alpha
beta = alpha + sigma
fBeta = beta ** 3 + 1

if fAlpha <= fBeta:
    b = beta
else:
    a = alpha

x = (a + b) / 2
f = x ** 3 - x
print(x, f)
```
